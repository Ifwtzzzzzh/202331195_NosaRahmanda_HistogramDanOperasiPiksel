## IMPORT LIBRARY

Library cv2 atau bisa disebut sebagai OpenCV kita gunakan untuk menganalisis dan memanipulasi gambar beserta video. Matplotlib.pyplot lebih berfokus pada visualisasi atau menampilkan bentuk data.

```python
import cv2
import matplotlib.pyplot as plt
```

## MEMBACA GAMBAR

Untuk membaca gambar bernama "porsche_cayman.jpg" saya menggunakan fungsi cv2.imread(). Kemudian, gambar tersebut ditampilkan dalam sebuah jendela dengan judul "Gambar" melalui fungsi cv2.imshow(). Fungsi cv2.waitKey(3000) akan menahan tampilan jendela selama 3000 milidetik (3 detik) atau hingga ada tombol keyboard ditekan. Terakhir, cv2.destroyAllWindows() digunakan untuk menutup semua jendela OpenCV yang telah dibuka.

```python
img = cv2.imread("porsche_cayman.jpg")
cv2.imshow("Gambar", img)
cv2.waitKey(3000)
cv2.destroyAllWindows()
```

## MELIHAT ATRIBUT GAMBAR

Kode print(img.shape) saya digunakan untuk mencetak dimensi dari array NumPy yang direpresentasikan oleh variabel img.

```python
print(img.shape)
```

## GRAYSCALE DENGAN OPENCV

Syntax cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) berfungsi untuk mengubah citra berwarna (img) dari format BGR (Blue, Green, Red) menjadi citra skala abu-abu. 

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imshow("Grayscale", gray)
cv2.waitKey(3000)
cv2.destroyAllWindows()
print(gray.shape)
```

## GRAYSCALE TO BINARY DENGAN OPENCV

Fungsi ini digunakan untuk melakukan segmentasi citra ke dalam dua level intensitas berdasarkan nilai threshold yang ditentukan. Dalam kode (tresh, binary) = cv2.threshold(gray, 128, 255, cv2.THRESH_BINARY), citra grayscale yang disimpan dalam variabel gray akan diproses.

```python
(tresh, binary) = cv2.threshold(gray, 128, 255, cv2.THRESH_BINARY)
cv2.imshow("Biner", binary)
cv2.waitKey(3000)
cv2.destroyAllWindows()
print(binary.shape)
```

## MENGGUNAKAN MATPLOTLIB

plt.imshow(img) adalah fungsi dari pustaka Matplotlib (khususnya modul pyplot) yang digunakan untuk menampilkan sebuah array (biasanya NumPy array) sebagai gambar. 

```python
plt.imshow(img)
```

kode rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) menggunakan fungsi cvtColor dari library OpenCV (cv2) untuk melakukan konversi ruang warna pada sebuah citra digital yang direpresentasikan oleh variabel img.

```python
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb)
```

# MENYIMPAN GAMBAR

Lalu, di sini saya akan menyimpan gambar dari variable gray dengan nama porsche_gray.png dan variable binary dengan nama porsche_black.png.

```python
cv2.imwrite('porsche_gray.png', gray)
cv2.imwrite('porsche_black.png', binary)
```
