# KONSEP DASAR CITRA

## IMPORT LIBRARY

```python
import cv2
```

```python
import matplotlib.pyplot as plt
```

## MEMBACA GAMBAR

```python
img = cv2.imread("lego.png")
cv2.imshow("Gambar", img)
cv2.waitKey(3000)
cv2.destroyAllWindows()
```

## MELIHAT ATRIBUT GAMBAR

```python
print(img.shape)
```

## GRAYSCALE DENGAN OPENCV

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imshow("Grayscale", gray)
cv2.waitKey(3000)
cv2.destroyAllWindows()
```

```python
print(gray.shape)
```

## GRAYSCALE TO BINARY DENGAN OPENCV

```python
(tresh, binary) = cv2.threshold(gray, 128, 255, cv2.THRESH_BINARY)
```

```python
cv2.imshow("Biner", binary)
cv2.waitKey(3000)
cv2.destroyAllWindows()
```

```python
print(binary.shape)
```

## MENGGUNAKAN MATPLOTLIB

```python
plt.imshow(img)
```

```python
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb)
```

# MENYIMPAN GAMBAR

```python
cv2.imwrite('Gambar_Abu.png', gray)
cv2.imwrite('Gambar_Hitam.png', binary)
```
