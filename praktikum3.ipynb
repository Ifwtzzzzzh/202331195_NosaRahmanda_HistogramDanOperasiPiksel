## IMPORT LIBRARY

Library cv2 kita gunakan untuk memanggil fungsi-fungsi pengolahan citra. Library matplotlib untuk visualisasi histogram. Dan library numpy untuk melakukan operasi data numerik menjadi array.

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

## MEMBACA DATA GAMBAR

Kodingan img = cv2.imread('porsche_cayman.jpg') membaca gambar bernama 'img_gray.jpg' dan menyimpannya dalam variabel img. Lalu, img.shape memberikan informasi dimensi gambar 

```python
img = cv2.imread('porsche_cayman.jpg')
img.shape
```

```python
[baris,kolom] = img.shape[:2]
```

## CONVERT BGR TO RGB

Kita akan mengubah ruang warna gambar dari BGR (Blue, Green, Red) yang umum digunakan OpenCV menjadi RGB (Red, Green, Blue).

```python
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

## MENAMPILKAN HISTOGRAM

Kita akan membuat dua set subplot yang menampilkan citra RGB dan histogramnya menggunakan Matplotlib dan OpenCV.

```python
fig, axs = plt.subplots(2,2, figsize = (15,5))

# Cara 1
axs[0,0].imshow(rgb)
axs[0,1].hist(rgb.ravel(),256,[0,256])

# Cara 2
hist = cv2.calcHist([rgb],[0],None,[256],[0,256])
axs[1,0].imshow(rgb)
axs[1,1].plot(hist)
plt.show()
```

## OPERASI PIKSEL UNTUK MENINGKATKAN KECERAHAN

Kode ini melakukan penambahan nilai konstan (beta) sebesar 40 ke setiap piksel citra berwarna (rgb) untuk meningkatkan kecerahannya, kemudian menampilkan citra asli dan citra yang telah ditingkatkan kecerahannya beserta histogramnya masing-masing dalam sebuah figure dengan subplots.

```python
beta = 40
citra_cerah = np.zeros((baris,kolom,3))

for x in range(baris):
    for y in range(kolom):
        gyx = rgb[x,y]+beta
        citra_cerah[x,y] = gyx

citra_cerah = citra_cerah.astype(np.uint8)
fig, axs = plt.subplots(2,2,figsize=(15,5))
axs[0,0].imshow(rgb)
axs[0,1].hist(rgb.ravel(),256,[0,256])
axs[1,0].imshow(citra_cerah)
axs[1,1].hist(citra_cerah.ravel(),256,[0,256])
plt.show()
```

## MEREGANGKAN KONTRAS

Kodingan ini meningkatkan kontras citra RGB dengan mengalikan setiap nilai piksel dengan faktor =1.2 dan kemudian menampilkan citra asli beserta histogramnya dan citra hasil peningkatan kontras beserta histogramnya dalam sebuah plot.

```python
alpha = 1.2
citra_kontras = np.zeros((baris,kolom,3))

for x in range(baris):
    for y in range(kolom):
        gyx = rgb[x,y]*alpha
        citra_kontras[x,y] = gyx

citra_kontras = citra_kontras.astype(np.uint8)
fig, axs = plt.subplots(2,2,figsize=(15,5))
axs[0,0].imshow(rgb)
axs[0,1].hist(rgb.ravel(),256,[0,256])
axs[1,0].imshow(citra_kontras)
axs[1,1].hist(citra_kontras.ravel(),256,[0,256])
plt.show()
```

## OPREASI GABUNGAN ANTARA MENINGKATKAN KECERAHAN DAN MEREGANGKAN KONTRAS

Kodingan ini melakukan penyesuaian kecerahan dan kontras citra berwarna dengan mengalikan setiap nilai piksel dengan faktor alpha dan menambahkan nilai beta, kemudian menampilkan citra asli dan hasilnya beserta histogramnya.

```python
beta = 10
alpha = 1.2
citra_hasil = np.zeros((baris,kolom,3))

for x in range(baris):
    for y in range(kolom):
        gyx = alpha*rgb[x,y]+beta
        citra_hasil[x,y] = gyx

citra_hasil = citra_hasil.astype(np.uint8)
fig, axs = plt.subplots(2,2,figsize=(15,5))
axs[0,0].imshow(rgb)
axs[0,1].hist(rgb.ravel(),256,[0,256])
axs[1,0].imshow(citra_hasil)
axs[1,1].hist(citra_hasil.ravel(),256,[0,256])
plt.show()
```

## MEMBALIK CITRA

Kodingan ini melakukan operasi linier pada intensitas piksel citra RGB dengan mengalikan setiap nilai piksel dengan faktor alpha dan menambahkan nilai beta, kemudian menampilkan citra asli dan hasil beserta histogramnya.

```python
citra_negatif = np.zeros((baris,kolom,3))

for x in range(baris):
    for y in range(kolom):
        gyx = alpha*rgb[x,y]+beta
        citra_negatif[x,y] = gyx

citra_negatif = citra_negatif.astype(np.uint8)
fig, axs = plt.subplots(2,2,figsize=(15,5))
axs[0,0].imshow(rgb)
axs[0,1].hist(rgb.ravel(),256,[0,256])
axs[1,0].imshow(citra_negatif)
axs[1,1].hist(citra_negatif.ravel(),256,[0,256])
plt.show()
```
