# IMPORT LIBRARY

Kode ini mengimpor tiga library Python yang umum digunakan untuk pemrosesan gambar dan visualisasi: `cv2` (OpenCV) untuk operasi gambar, `numpy` (np) untuk komputasi numerik, dan `matplotlib.pyplot` (plt) untuk menampilkan gambar dan plot. Bersama-sama, library ini memungkinkan manipulasi gambar, analisis data array, dan visualisasi data secara efisien.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# LOAD GAMBAR

Kode ini membaca gambar 'porsche_cayman.jpg' menggunakan OpenCV, lalu mengkonversi format warnanya dari BGR (standar OpenCV) ke RGB (standar matplotlib) agar ditampilkan dengan warna yang benar. Gambar kemudian ditampilkan menggunakan matplotlib dengan judul 'Gambar Asli' dan tanpa sumbu koordinat. Fungsi `plt.show()` menampilkan gambar dalam jendela terpisah.

```python
img = cv2.imread('porsche_cayman.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.title('Gambar Asli')
plt.axis('off')
plt.show()
```

# MEAN FILTER

Kode ini menerapkan **mean filter** (filter rata-rata) berukuran 5x5 pada citra menggunakan `cv2.blur()` untuk menghaluskan noise dengan mengganti setiap piksel dengan rata-rata nilai piksel tetangganya. Hasilnya ditampilkan menggunakan `matplotlib` dengan judul *"Hasil Filter Rata-rata (Mean)"* dan menghilangkan sumbu koordinat (`plt.axis('off')`). Operasi ini berguna untuk **denoising** atau **blurring** sederhana dalam pengolahan citra.

```python
mean_filtered = cv2.blur(img, (5, 5))
plt.imshow(mean_filtered)
plt.title('Hasil Filter Rata-rata (Mean)')
plt.axis('off')
plt.show()
```

# MEDIAN FILTER

Kode ini menerapkan filter median berukuran 5x5 pada gambar (`img`) menggunakan `cv2.medianBlur` untuk mengurangi noise (seperti salt-and-pepper) sambil mempertahankan tepi tajam. Hasilnya ditampilkan menggunakan `matplotlib` dengan judul "Hasil Filter Median" dan menghilangkan sumbu koordinat (`axis('off')`). Fungsi `plt.show()` menampilkan gambar yang telah difilter dalam sebuah window visual.

```python
median_filtered = cv2.medianBlur(img, 5)
plt.imshow(median_filtered)
plt.title('Hasil Filter Median')
plt.axis('off')
plt.show()
```

# EDGE DETECTION (SOBEL) FILTER

Kode ini mengubah citra berwarna (`img`) menjadi grayscale, lalu menghitung gradien tepi menggunakan operator Sobel pada arah horizontal (`sobel_x`) dan vertikal (`sobel_y`). Gradien ini digabungkan menggunakan `cv2.magnitude()` untuk mendapatkan tepi yang lebih jelas, yang kemudian ditampilkan dalam plot grayscale menggunakan `matplotlib`. Hasilnya adalah visualisasi deteksi tepi Sobel dengan judul "Deteksi Tepi (Sobel)" tanpa sumbu koordinat.

```python
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
sobel_x = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=5)
sobel_y = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=5)
edges = cv2.magnitude(sobel_x, sobel_y)

plt.imshow(edges, cmap='gray')
plt.title('Deteksi Tepi (Sobel)')
plt.axis('off')
plt.show()
```

# FILTER 2D

Kode ini memuat gambar 'boneka.jpg', mengubah format warnanya dari BGR ke RGB, lalu menerapkan filter konvolusi menggunakan kernel rata-rata 5x5 untuk efek blurring. Selanjutnya, menampilkan dua gambar secara berdampingan: gambar asli dan hasil filter menggunakan `matplotlib`, dengan judul dan pengaturan tampilan yang sesuai. Proses ini menunjukkan operasi dasar pemrosesan citra untuk menghaluskan gambar.

```python
sapi = cv2.imread('boneka.jpg')
sapi = cv2.cvtColor(sapi, cv2.COLOR_BGR2RGB)
kernel = np.ones((5,5), np.float32) / 25
filtered_sapi = cv2.filter2D(sapi, -1, kernel)

plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.imshow(sapi)
plt.title('Boneka Sapi - Asli')
plt.axis('off')
plt.subplot(1,2,2)
plt.imshow(filtered_sapi)
plt.title('Boneka Sapi - Filter 2D')
plt.axis('off')
plt.show()
```
